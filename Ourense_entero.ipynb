```python
import folium
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/manurial/Desktop/ourense.csv')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
subset_of_df=df.sample(n=194)
```

```python
some_map=folium.Map(location=[subset_of_df['COORDENADA_X'].mean(),
                     subset_of_df['COORDENADA_Y'].mean()],
                     zoom_start=10)
```

```python
for row in subset_of_df.itertuples():
    my_string = "OU-Centro",row.Código,row.COORDENADA_X,row.COORDENADA_Y
    some_map.add_child(folium.Marker(location=[row.COORDENADA_X,row.COORDENADA_Y],popup=(my_string)))
```

```python
some_map
```

```python
from folium.plugins import MarkerCluster
```

```python
some_map_2=folium.Map(location=[subset_of_df['COORDENADA_X'].mean(),
                     subset_of_df['COORDENADA_Y'].mean()],
                     zoom_start=10)
```

```python
for row in subset_of_df.itertuples():
    mc.add_child(folium.Marker(location=[row.COORDENADA_X,row.COORDENADA_Y],popup=row.Código))
```
