```python
import folium #importamos la librería folium de graficado geoposicional
```

```python
import pandas as pd #importamos la librería pandas de manejo de dataframes-matrices de datos
```

```python
df=pd.read_csv('/Users/manurial/Desktop/coru.csv') #cargamos el fichero de centros de a coruña en una matriz
```

```python
df.columns = df.columns.str.strip() #transformamos el dataframe en registros para leerlo como una base de datos
```

```python
df.head() #mostramos los 5 primeros registros
```

```python
df.tail() #mostramos los 5 últimos registros
```

```python
subset_of_df=df.sample(n=300) #seleccionamos un subconjunto de datos de la base de datos, los 300 primeros elementos
```

```python
some_map=folium.Map(location=[subset_of_df['X'].mean(),
                     subset_of_df['Y'].mean()],
                     zoom_start=10)
#definimos el tipo de mapa determinando el zoom, y las distancias promedio de cada una de las mismas
```

```python
for row in subset_of_df.itertuples():
    my_string = "Centro",row.ID,row.X,row.Y
    some_map.add_child(folium.Marker(location=[row.X,row.Y],popup=(my_string)))
#mostramos en el mapa los marcadores correspondientes a los centros escolares
```

```python
some_map
```

```python
from folium.plugins import MarkerCluster
#importamos un plugin que permitirá manejar los marcadores agrupándolos
```

```python
some_map_2=folium.Map(location=[subset_of_df['X'].mean(),
                     subset_of_df['Y'].mean()],
                     zoom_start=10)
#definimos un segundo mapa con iguales características que el anterior
```

```python
mc=MarkerCluster()
#definimos un nuevo manejador de marcadores agrupados
```

```python
for row in subset_of_df.itertuples():
    mc.add_child(folium.Marker(location=[row.X,row.Y],popup=row.ID))
#mostramos dinámicamente los marcadores agrupados
```

```python
some_map_2.add_child(mc)
#graficamos el mapa correspondiente
```
