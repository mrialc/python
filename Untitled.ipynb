```python
import folium
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/manurial/Desktop/coru.csv')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df.tail()
```

```python
subset_of_df=df.sample(n=300)
```

```python
some_map=folium.Map(location=[subset_of_df['X'].mean(),
                     subset_of_df['Y'].mean()],
                     zoom_start=10)
```

```python
for row in subset_of_df.itertuples():
    some_map.add_child(folium.Marker(location=[row.X,row.Y],popup=row.ID))
```

```python
some_map
```

```python
from folium.plugins import MarkerCluster
```

```python
some_map_2=folium.Map(location=[subset_of_df['X'].mean(),
                     subset_of_df['Y'].mean()],
                     zoom_start=10)
```
